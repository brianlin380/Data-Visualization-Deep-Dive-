```python
import pandas as pd
from matplotlib import colormaps
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("building_inventory.csv", na_values={'Square Footage': 0, 
                                                    'Year Constructed':0,
                                                    'Year Acquired':0}) 
df
```

```python
plt.scatter(x=df['Total Floors'], y=np.sqrt(df['Square Footage'])) 
plt.xlabel("Total Floors") 
plt.ylabel("Square Footage") 
plt.show()
```

```python
top_agencies = df['Agency Name'].value_counts()[:12]
top_agencies
```

```python
plt.bar(top_agencies.index, top_agencies.values, color=colormaps['Pastel1'].colors)
plt.ylabel("Number of Buildings")
plt.xticks(rotation=90)
plt.title("Buildings in Illinois Agencies")
plt.show()
```
